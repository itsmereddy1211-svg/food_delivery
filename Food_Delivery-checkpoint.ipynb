```python
import sqlite3
```

```python
orders_df = pd.read_csv("orders.csv")
final_df = orders_df.copy()
df = final_df.copy()
df.head()
```

```python
users_df=pd.read_json("users.json")
users_df.head()
users_df.head()
```

```python
conn = sqlite3.connect(":memory:")
with open("restaurants.sql", "r") as file:
    sql_script = file.read()
conn.executescript(sql_script)
restaurants_df = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants_df.head()
```

```python
print("orders exists:", "orders" in globals())
print("users exists:", "users" in globals())
print("restaurants exists:", "restaurants" in globals())
```

```python
merged_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)
final_df = pd.merge(
    merged_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.shape
final_df.info()
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
##MCQS
df[df["membership"] == "Gold"] \
  .groupby("city")["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
df.groupby("cuisine")["total_amount"] \
  .mean() \
  .sort_values(ascending=False)
```

```python
user_spend = df.groupby("user_id")["total_amount"].sum()
user_spend[user_spend > 1000].count()
```

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ["3.0–3.5", "3.6–4.0", "4.1–4.5", "4.6–5.0"]
df["rating_range"] = pd.cut(
    df["rating"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
df.groupby("rating_range", observed=True)["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
df[df["membership"] == "Gold"] \
  .groupby("city")["total_amount"] \
  .mean() \
  .sort_values(ascending=False)
```

```python
df.groupby("cuisine")["restaurant_id"] \
  .nunique() \
  .sort_values()
df.groupby("cuisine")["total_amount"].sum()
```

```python
gold_pct = (
    df[df["membership"] == "Gold"].shape[0] / df.shape[0]
) * 100
round(gold_pct)
```

```python
restaurant_stats = df.groupby("restaurant_name_x").agg(
    avg_order_value=("total_amount", "mean"),
    total_orders=("order_id", "count")
)

restaurant_stats[
    restaurant_stats["total_orders"] < 20
].sort_values("avg_order_value", ascending=False)
```

```python
df.groupby(["membership", "cuisine"])["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)

df["quarter"] = df["order_date"].dt.quarter
df.groupby("quarter")["total_amount"] \
  .sum() \
  .sort_values(ascending=False)
```
